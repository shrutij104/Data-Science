```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set default style
sns.set(style="darkgrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.shape  # Rows and columns
df.columns  # Feature names
df.info()  # Data types and nulls
df.describe(include='all')  # Summary stats
```

```python
df.isnull().sum()

# Visualize missing values
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
```

```python
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
```

```python
for col in ['Sex', 'Pclass', 'Embarked']:
    sns.countplot(x=col, data=df)
    plt.title(f'{col} Distribution')
    plt.show()
```

```python
df['Age'].hist(bins=30)
plt.title('Age Distribution')

sns.histplot(df['Fare'], bins=40, kde=True)
plt.title('Fare Distribution')
```

```python
sns.boxplot(x='Age', data=df)
plt.title('Boxplot - Age')

sns.boxplot(x='Fare', data=df)
plt.title('Boxplot - Fare')
```

```python
# Sex vs Survival
sns.barplot(x='Sex', y='Survived', data=df)

# Class vs Survival
sns.barplot(x='Pclass', y='Survived', data=df)

# Age vs Survival
sns.kdeplot(data=df[df['Survived']==1]['Age'], label='Survived', shade=True)
sns.kdeplot(data=df[df['Survived']==0]['Age'], label='Not Survived', shade=True)
plt.title('Age Distribution by Survival')
plt.legend()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
```
